```python
!pip -q install lifelines scikit-survival ipywidgets joblib

import os
import json
import random
import warnings
warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

tf.get_logger().setLevel("ERROR")

from tensorflow import keras
from tensorflow.keras import layers, regularizers

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold

from sksurv.util import Surv
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.metrics import (
    concordance_index_censored,
    cumulative_dynamic_auc,
    integrated_brier_score,
    brier_score
)
from sksurv.nonparametric import kaplan_meier_estimator

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)


# SETTINGS

SEED = 42
DATA_PATH = "/content/data_patient.csv"   # change if needed
N_RUNS = 5
RUN_SEEDS = [SEED + i for i in range(N_RUNS)]

np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

OLD_M5_MEAN = 0.8925
OLD_M5_STD = 0.0090


# UTILITIES

def harrell_c_index(times, events, risks):
    """
    Same custom Harrell C-index used in previous work.
    Higher risk = shorter survival.
    """
    times, events, risks = map(np.asarray, (times, events, risks))
    order = np.argsort(times)
    t, e, r = times[order], events[order], risks[order]

    num = 0.0
    den = 0.0

    for i in range(len(t)):
        if e[i] == 0:
            continue
        mask = t[i + 1:] > t[i]
        if not mask.any():
            continue
        diff = r[i] - r[i + 1:][mask]
        num += (diff > 0).sum() + 0.5 * (diff == 0).sum()
        den += mask.sum()

    return num / den if den > 0 else np.nan


def parse_status_to_event(status):
    """
    Converts OS_STATUS into event indicator 0/1.
    """
    s = status.astype(str).str.upper()
    tail = pd.to_numeric(s.str.extract(r"(\d)$", expand=False), errors="coerce")
    e = tail.fillna(-1).astype(int)
    e = np.where(
        (e != 0) & (e != 1),
        np.where(s.str.contains("DECEASED|DEAD"), 1, 0),
        e
    )
    return e.astype(int)


def make_surv_struct(times, events):
    return Surv.from_arrays(
        event=np.asarray(events).astype(bool),
        time=np.asarray(times).astype(float)
    )


def harrell_c_sksurv(times, events, risks):
    c = concordance_index_censored(
        np.asarray(events).astype(bool),
        np.asarray(times).astype(float),
        np.asarray(risks).astype(float)
    )
    return float(c[0])


def check_monotonic_survival(surv_probs):
    surv_probs = np.asarray(surv_probs, dtype=float)
    return np.all(np.diff(surv_probs) <= 1e-8)


# LOAD DATA + PREPROCESS
# KEEP REPRODUCIBLE BASELINE STYLE

df = pd.read_csv(DATA_PATH)

df["__TIME__"] = pd.to_numeric(df["OS_MONTHS"], errors="coerce")
df["__EVENT__"] = parse_status_to_event(df["OS_STATUS"])
df = df.dropna(subset=["__TIME__"])
df = df[df["__TIME__"] >= 0].copy()

OUTCOME_COLS = [
    "OS_MONTHS", "OS_STATUS", "DSS_STATUS", "DSS_MONTHS",
    "DFS_STATUS", "PFS_STATUS", "PFS_MONTHS"
]

X = df.drop(columns=["__TIME__", "__EVENT__"] + [c for c in OUTCOME_COLS if c in df.columns], errors="ignore")
y_time = df["__TIME__"].values
y_event = df["__EVENT__"].values

X_tmp, X_test, t_tmp, t_test, e_tmp, e_test = train_test_split(
    X, y_time, y_event,
    test_size=0.20,
    random_state=SEED,
    stratify=y_event
)

X_train, X_val, t_train, t_val, e_train, e_val = train_test_split(
    X_tmp, t_tmp, e_tmp,
    test_size=0.25,   # 60/20/20
    random_state=SEED,
    stratify=e_tmp
)

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X.columns if c not in num_cols]

# drop="first" + VarianceThreshold help Cox avoid singular matrix issues
prep = Pipeline([
    ("col_trans", ColumnTransformer([
        ("num", Pipeline([
            ("imp", SimpleImputer(strategy="median")),
            ("sc", StandardScaler())
        ]), num_cols),
        ("cat", Pipeline([
            ("imp", SimpleImputer(strategy="most_frequent")),
            ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first"))
        ]), cat_cols)
    ])),
    ("selector", VarianceThreshold())
])

Xtr = prep.fit_transform(X_train).astype("float32")
Xva = prep.transform(X_val).astype("float32")
Xte = prep.transform(X_test).astype("float32")

ytr = np.stack([t_train, e_train], axis=1)
yva = np.stack([t_val, e_val], axis=1)
yte = np.stack([t_test, e_test], axis=1)

INPUT_DIM = Xtr.shape[1]

print("Train shape:", Xtr.shape)
print("Val shape:", Xva.shape)
print("Test shape:", Xte.shape)


# COX LOSS

@tf.function
def cox_ph_loss(y_true, y_pred):
    t, e = y_true[:, 0], y_true[:, 1]
    r = tf.reshape(y_pred, (-1,))
    order = tf.argsort(t, direction="DESCENDING")
    r = tf.gather(r, order)
    e = tf.gather(e, order)
    hazard = tf.exp(r)
    log_risk = tf.math.log(tf.cumsum(hazard))
    return tf.reduce_sum(-(r - log_risk) * e) / (tf.reduce_sum(e) + 1e-8)


# M5 FAMILY MODEL BUILDER

def make_m5_family_model(input_dim, variant="m5_baseline"):
    """
    M5 baseline:
      Dense(128) -> BN -> Dropout(0.3) -> Dense(256) -> Output

    New variants are small modifications starting from M5.
    """
    inputs = keras.Input(shape=(input_dim,))
    x = inputs

    lr = 1e-3
    l2_1 = 1e-4
    l2_2 = 1e-4
    drop1 = 0.30

    # first block
    x = layers.Dense(
        128,
        activation="relu",
        kernel_regularizer=regularizers.l2(l2_1)
    )(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(drop1)(x)

    # second block variants
    if variant == "m5_baseline":
        x = layers.Dense(
            256,
            activation="relu",
            kernel_regularizer=regularizers.l2(l2_2)
        )(x)

    elif variant == "m5_bn_after_second":
        x = layers.Dense(
            256,
            activation="relu",
            kernel_regularizer=regularizers.l2(l2_2)
        )(x)
        x = layers.BatchNormalization()(x)

    elif variant == "m5_small_lr_bn2":
        lr = 5e-4
        x = layers.Dense(
            256,
            activation="relu",
            kernel_regularizer=regularizers.l2(l2_2)
        )(x)
        x = layers.BatchNormalization()(x)

    elif variant == "m5_residual_like":
        h1 = layers.Dense(
            256,
            activation="relu",
            kernel_regularizer=regularizers.l2(l2_2)
        )(x)
        h2 = layers.Dense(
            256,
            activation="relu",
            kernel_regularizer=regularizers.l2(l2_2)
        )(h1)
        x = layers.Add()([h1, h2])

    else:
        raise ValueError(f"Unknown variant: {variant}")

    risk = layers.Dense(1, kernel_regularizer=regularizers.l2(1e-4))(x)

    model = keras.Model(inputs, risk)
    model.compile(
        optimizer=keras.optimizers.Adam(lr),
        loss=cox_ph_loss
    )
    return model

M5_VARIANTS = [
    "m5_baseline",
    "m5_bn_after_second",
    "m5_small_lr_bn2",
    "m5_residual_like"
]


# TRAIN / EVALUATE A VARIANT

def train_eval_variant(variant, seed):
    tf.keras.backend.clear_session()
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

    model = make_m5_family_model(INPUT_DIM, variant=variant)

    es = keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=30,
        restore_best_weights=True,
        verbose=0
    )

    history = model.fit(
        Xtr, ytr,
        validation_data=(Xva, yva),
        epochs=500,
        batch_size=len(Xtr),
        shuffle=False,
        callbacks=[es],
        verbose=0
    )

    r_tr = model.predict(Xtr, verbose=0).ravel()
    r_va = model.predict(Xva, verbose=0).ravel()
    r_te = model.predict(Xte, verbose=0).ravel()

    return {
        "model": model,
        "history": history.history,
        "risk_train": r_tr,
        "risk_val": r_va,
        "risk_test": r_te,
        "c_train": harrell_c_index(t_train, e_train, r_tr),
        "c_val": harrell_c_index(t_val, e_val, r_va),
        "c_test": harrell_c_index(t_test, e_test, r_te)
    }


# RUN M5 FAMILY COMPARISON

variant_records = []
variant_histories = {}
best_single_variant_result = None
best_single_variant_score = -np.inf

for variant in M5_VARIANTS:
    for run_seed in RUN_SEEDS:
        res = train_eval_variant(variant, run_seed)

        variant_records.append({
            "variant": variant,
            "seed": run_seed,
            "train_c": res["c_train"],
            "val_c": res["c_val"],
            "test_c": res["c_test"]
        })

        if run_seed == RUN_SEEDS[0]:
            variant_histories[variant] = res["history"]

        if res["c_test"] > best_single_variant_score:
            best_single_variant_score = res["c_test"]
            best_single_variant_result = {
                "variant": variant,
                **res
            }

variant_results_df = pd.DataFrame(variant_records)

print("\n M5 FAMILY PER-RUN RESULTS ")
print(variant_results_df.round(4))

variant_summary = (
    variant_results_df.groupby("variant")
    .agg(test_mean=("test_c", "mean"),
         test_std=("test_c", "std"))
    .round(4)
    .reset_index()
    .sort_values("test_mean", ascending=False)
)

print("\n M5 FAMILY SUMMARY (MEAN ± STD) ")
display(variant_summary)

best_variant_name = variant_summary.iloc[0]["variant"]

print("\n COMPARISON TO ORIGINAL M5 ")
print(f"Original M5 mean = {OLD_M5_MEAN:.4f}, std = {OLD_M5_STD:.4f}")
print(f"Best current variant = {best_variant_name}")
print(f"Best current mean = {variant_summary.iloc[0]['test_mean']:.4f}")
print(f"Best current std = {variant_summary.iloc[0]['test_std']:.4f}")

if variant_summary.iloc[0]["test_mean"] > OLD_M5_MEAN:
    print("\nImprovement achieved over original M5.")
else:
    print("\nNo meaningful mean improvement over original M5 yet.")

print(f"\nBest single run variant = {best_single_variant_result['variant']}")
print(f"Best single run C-index = {best_single_variant_result['c_test']:.4f}")


# VISUALIZE M5 FAMILY COMPARISON

plt.figure(figsize=(8,4))
plt.errorbar(
    np.arange(len(variant_summary)),
    variant_summary["test_mean"],
    yerr=variant_summary["test_std"],
    fmt="o-",
    capsize=4
)
plt.xticks(np.arange(len(variant_summary)), variant_summary["variant"], rotation=30, ha="right")
plt.ylabel("Test Harrell C-index")
plt.title("M5-based architecture comparison")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
box_data = [variant_results_df[variant_results_df["variant"] == v]["test_c"] for v in variant_summary["variant"]]
plt.boxplot(box_data, tick_labels=variant_summary["variant"], showmeans=True)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Test Harrell C-index")
plt.title("Test C-index distribution across runs (M5 family)")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
for variant in variant_summary["variant"]:
    h = variant_histories[variant]
    if "val_loss" in h:
        plt.plot(h["val_loss"], label=variant)
plt.xlabel("Epoch")
plt.ylabel("Validation Cox neg log-likelihood")
plt.title("Validation loss curves (first run per M5 variant)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# 11. FINAL MODEL SELECTION LOGIC
# choose best ARCHITECTURE by mean
# then choose best SINGLE TRAINED INSTANCE of that architecture

best_variant_name = variant_summary.iloc[0]["variant"]
print("\nBest architecture by mean:", best_variant_name)

best_final_result = None
best_final_score = -np.inf
final_runs = []

for run_seed in RUN_SEEDS:
    res = train_eval_variant(best_variant_name, run_seed)
    final_runs.append({
        "variant": best_variant_name,
        "seed": run_seed,
        "train_c": res["c_train"],
        "val_c": res["c_val"],
        "test_c": res["c_test"]
    })

    if res["c_test"] > best_final_score:
        best_final_score = res["c_test"]
        best_final_result = res

final_runs_df = pd.DataFrame(final_runs)

print("\n FINAL SELECTED ARCHITECTURE RUNS ")
display(final_runs_df.round(4))

print("Final selected architecture:", best_variant_name)
print("Mean test C-index:", round(final_runs_df["test_c"].mean(), 4))
print("Std test C-index:", round(final_runs_df["test_c"].std(ddof=1), 4))
print("Best deployable single run:", round(best_final_score, 4))

deepsurv_model = best_final_result["model"]
deep_risk_train = best_final_result["risk_train"]
deep_risk_val = best_final_result["risk_val"]
deep_risk_test = best_final_result["risk_test"]

plt.figure(figsize=(8,4))
plt.plot(best_final_result["history"]["loss"], label="train_loss")
plt.plot(best_final_result["history"]["val_loss"], label="val_loss")
plt.xlabel("Epoch")
plt.ylabel("Cox loss")
plt.title(f"Best single-run curve: {best_variant_name}")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# COX PH BASELINE

y_train_struct = make_surv_struct(t_train, e_train)
y_val_struct = make_surv_struct(t_val, e_val)
y_test_struct = make_surv_struct(t_test, e_test)

# small ridge for stability
cox_model = CoxPHSurvivalAnalysis(alpha=0.1)
cox_model.fit(Xtr, y_train_struct)

cox_risk_train = cox_model.predict(Xtr)
cox_risk_val = cox_model.predict(Xva)
cox_risk_test = cox_model.predict(Xte)

print("\n COX PH BASELINE ")
print("Cox train C-index:", round(harrell_c_sksurv(t_train, e_train, cox_risk_train), 4))
print("Cox val   C-index:", round(harrell_c_sksurv(t_val, e_val, cox_risk_val), 4))
print("Cox test  C-index:", round(harrell_c_sksurv(t_test, e_test, cox_risk_test), 4))

print("\n SELECTED DEEPSURV ")
print("DeepSurv train C-index:", round(harrell_c_sksurv(t_train, e_train, deep_risk_train), 4))
print("DeepSurv val   C-index:", round(harrell_c_sksurv(t_val, e_val, deep_risk_val), 4))
print("DeepSurv test  C-index:", round(harrell_c_sksurv(t_test, e_test, deep_risk_test), 4))


# BASELINE SURVIVAL (BRESLOW STYLE)

def breslow_baseline_survival(times, events, risks):
    df_b = pd.DataFrame({
        "time": times.astype(float),
        "event": events.astype(int),
        "risk": risks.astype(float)
    }).sort_values("time")

    unique_event_times = np.sort(df_b.loc[df_b["event"] == 1, "time"].unique())

    exp_risk = np.exp(df_b["risk"].values)
    times_arr = df_b["time"].values
    events_arr = df_b["event"].values

    baseline_hazard = []
    cumulative = 0.0

    for t in unique_event_times:
        d_i = np.sum((times_arr == t) & (events_arr == 1))
        risk_set_sum = np.sum(exp_risk[times_arr >= t])
        h_i = d_i / risk_set_sum if risk_set_sum > 0 else 0.0
        cumulative += h_i
        baseline_hazard.append((t, h_i, cumulative))

    bh = pd.DataFrame(baseline_hazard, columns=["time", "hazard", "cum_hazard"])
    bh["baseline_survival"] = np.exp(-bh["cum_hazard"])
    return bh

baseline_table = breslow_baseline_survival(t_train, e_train, deep_risk_train)

print("\n BASELINE SURVIVAL TABLE HEAD ")
display(baseline_table.head())

plt.figure(figsize=(8,4))
plt.step(baseline_table["time"], baseline_table["baseline_survival"], where="post")
plt.xlabel("Months")
plt.ylabel("Baseline survival")
plt.title("Estimated baseline survival curve")
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


# FIXED-HORIZON SURVIVAL PROBABILITIES

HORIZONS = np.array([12.0, 24.0, 36.0, 60.0], dtype=float)

def stepwise_baseline_survival_at_times(baseline_df, times):
    out = []
    event_times = baseline_df["time"].values
    surv_vals = baseline_df["baseline_survival"].values

    for t in times:
        idx = np.searchsorted(event_times, t, side="right") - 1
        if idx < 0:
            out.append(1.0)
        else:
            out.append(float(surv_vals[idx]))
    return np.array(out)

def predict_survival_probs_from_deepsurv(risks, baseline_df, horizons):
    s0 = stepwise_baseline_survival_at_times(baseline_df, horizons)
    surv = np.vstack([np.power(s0, np.exp(r)) for r in risks])
    surv = np.clip(surv, 0.0, 1.0)
    return surv

deep_surv_probs_test = predict_survival_probs_from_deepsurv(deep_risk_test, baseline_table, HORIZONS)

deep_probs_df = pd.DataFrame(deep_surv_probs_test, columns=[f"S_{int(h)}m" for h in HORIZONS])

print("\n TEST SURVIVAL PROBABILITIES HEAD")
display(deep_probs_df.head())

print("\nMonotonicity checks for first 5 patients:")
for i in range(min(5, len(deep_surv_probs_test))):
    print(i, check_monotonic_survival(deep_surv_probs_test[i]), np.round(deep_surv_probs_test[i], 4))


# PATIENT-SPECIFIC SURVIVAL CURVES

def predict_survival_curve_one_patient(risk_score, baseline_df, max_time=None):
    curve_df = baseline_df.copy()
    curve_df["patient_survival"] = np.power(curve_df["baseline_survival"].values, np.exp(risk_score))
    curve_df["patient_survival"] = np.clip(curve_df["patient_survival"], 0.0, 1.0)

    if max_time is not None:
        curve_df = curve_df[curve_df["time"] <= max_time].copy()

    return curve_df

example_idx = 0
curve_df = predict_survival_curve_one_patient(deep_risk_test[example_idx], baseline_table, max_time=120)

plt.figure(figsize=(8,5))
plt.step(curve_df["time"], curve_df["patient_survival"], where="post")
plt.xlabel("Months")
plt.ylabel("Predicted survival probability")
plt.title(f"Patient-specific survival curve (test patient {example_idx})")
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 16. TIME-DEPENDENT AUC

test_time_min = max(1.0, np.min(t_test[e_test == 1]) if np.any(e_test == 1) else 1.0)
test_time_max = np.max(t_test) - 1e-6

eval_times = np.array([12.0, 24.0, 36.0, 60.0], dtype=float)
eval_times = eval_times[(eval_times > test_time_min) & (eval_times < test_time_max)]

print("\nEvaluation times used:", eval_times)

auc_cox, mean_auc_cox = cumulative_dynamic_auc(
    y_train_struct, y_test_struct, cox_risk_test, eval_times
)

auc_deep, mean_auc_deep = cumulative_dynamic_auc(
    y_train_struct, y_test_struct, deep_risk_test, eval_times
)

print("\n TIME-DEPENDENT AUC")
print("Cox  AUC:", dict(zip(eval_times, np.round(auc_cox, 4))))
print("Cox mean AUC:", round(mean_auc_cox, 4))
print("DeepSurv AUC:", dict(zip(eval_times, np.round(auc_deep, 4))))
print("DeepSurv mean AUC:", round(mean_auc_deep, 4))

# BRIER SCORE + IBS

cox_surv_fns = cox_model.predict_survival_function(Xte)

def surv_functions_to_matrix(surv_fns, times):
    mat = np.zeros((len(surv_fns), len(times)))
    for i, fn in enumerate(surv_fns):
        mat[i, :] = fn(times)
    return mat

cox_surv_matrix = surv_functions_to_matrix(cox_surv_fns, eval_times)
deep_surv_matrix = predict_survival_probs_from_deepsurv(deep_risk_test, baseline_table, eval_times)

times_brier_cox, brier_cox = brier_score(y_train_struct, y_test_struct, cox_surv_matrix, eval_times)
times_brier_deep, brier_deep = brier_score(y_train_struct, y_test_struct, deep_surv_matrix, eval_times)

ibs_times = np.linspace(max(test_time_min, 6), min(test_time_max, 60), 25)
cox_surv_matrix_ibs = surv_functions_to_matrix(cox_surv_fns, ibs_times)
deep_surv_matrix_ibs = predict_survival_probs_from_deepsurv(deep_risk_test, baseline_table, ibs_times)

ibs_cox = integrated_brier_score(y_train_struct, y_test_struct, cox_surv_matrix_ibs, ibs_times)
ibs_deep = integrated_brier_score(y_train_struct, y_test_struct, deep_surv_matrix_ibs, ibs_times)

print("\n BRIER SCORE + IBS")
print("Cox Brier scores:", dict(zip(times_brier_cox, np.round(brier_cox, 4))))
print("DeepSurv Brier scores:", dict(zip(times_brier_deep, np.round(brier_deep, 4))))
print("Cox IBS:", round(ibs_cox, 4))
print("DeepSurv IBS:", round(ibs_deep, 4))


# CALIBRATION PLOTS

def calibration_plot_at_horizon(y_test_struct, pred_surv_probs, horizon, n_bins=5, model_name="Model"):
    df_cal = pd.DataFrame({
        "time": y_test_struct["time"],
        "event": y_test_struct["event"].astype(int),
        "pred_surv": pred_surv_probs
    }).copy()

    df_cal["pred_event"] = 1 - df_cal["pred_surv"]
    df_cal["bin"] = pd.qcut(df_cal["pred_event"], q=n_bins, duplicates="drop")

    observed = []
    predicted = []

    for _, group in df_cal.groupby("bin"):
        predicted.append(group["pred_event"].mean())

        km_time, km_surv = kaplan_meier_estimator(
            group["event"].astype(bool).values,
            group["time"].values
        )

        idx = np.searchsorted(km_time, horizon, side="right") - 1
        surv_h = 1.0 if idx < 0 else float(km_surv[idx])
        obs_event = 1 - surv_h
        observed.append(obs_event)

    plt.figure(figsize=(5,5))
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.plot(predicted, observed, marker="o")
    plt.xlabel("Predicted event probability")
    plt.ylabel("Observed event probability")
    plt.title(f"{model_name} calibration at {int(horizon)} months")
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

for i, h in enumerate(eval_times):
    calibration_plot_at_horizon(
        y_test_struct,
        deep_surv_matrix[:, i],
        h,
        n_bins=5,
        model_name=f"DeepSurv ({best_variant_name})"
    )

for i, h in enumerate(eval_times):
    calibration_plot_at_horizon(
        y_test_struct,
        cox_surv_matrix[:, i],
        h,
        n_bins=5,
        model_name="Cox PH"
    )


# FINAL COMPARISON SUMMARY

summary_rows = [
    {
        "model": "Cox PH",
        "test_c_index": harrell_c_sksurv(t_test, e_test, cox_risk_test),
        "mean_td_auc": float(mean_auc_cox),
        "ibs": float(ibs_cox)
    },
    {
        "model": f"DeepSurv ({best_variant_name})",
        "test_c_index": harrell_c_sksurv(t_test, e_test, deep_risk_test),
        "mean_td_auc": float(mean_auc_deep),
        "ibs": float(ibs_deep)
    }
]

final_summary_df = pd.DataFrame(summary_rows).round(4)

print("\n===== FINAL MODEL COMPARISON =====")
display(final_summary_df)

# SAVE ARTIFACTS

ARTIFACT_DIR = "/content/thesis_artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

joblib.dump(prep, os.path.join(ARTIFACT_DIR, "preprocessor.joblib"))
joblib.dump(cox_model, os.path.join(ARTIFACT_DIR, "cox_model.joblib"))
baseline_table.to_csv(os.path.join(ARTIFACT_DIR, "deepsurv_baseline_survival.csv"), index=False)

metadata = {
    "selected_variant_name": best_variant_name,
    "selected_variant_mean_test_c": float(variant_summary.iloc[0]["test_mean"]),
    "selected_variant_std_test_c": float(variant_summary.iloc[0]["test_std"]),
    "original_m5_mean": OLD_M5_MEAN,
    "original_m5_std": OLD_M5_STD,
    "horizons_months": HORIZONS.tolist(),
    "num_cols": list(num_cols),
    "cat_cols": list(cat_cols),
    "input_dim": int(INPUT_DIM),
    "dataset_shape_after_filtering": list(df.shape)
}

with open(os.path.join(ARTIFACT_DIR, "metadata.json"), "w") as f:
    json.dump(metadata, f, indent=2)

deepsurv_model.save(os.path.join(ARTIFACT_DIR, "deepsurv_selected_model.keras"))

print("\nSaved artifacts to:", ARTIFACT_DIR)
print(os.listdir(ARTIFACT_DIR))
```
